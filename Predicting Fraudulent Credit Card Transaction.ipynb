#### <font color='steelblue'>SITUATIONER:</font> Company is interested to implement a machine learning model that will identify fraudulent transactions. The data set provided contains 30 features and 284,807 records. The V-Columns represent components from Principal Component Analysis, the variable labeled Amount unchanged, the variable Class is although numerical (ready for modelling) has 1 which refers to fraudulent transaction whereas 0 is non-fraudulent. 

```python
# import needed libraries
import pandas as pd
import numpy as np
import researchpy as rp
from sklearn.preprocessing import StandardScaler
```

```python
# load needed data from the CSV file
df = pd.read_csv("creditcard.csv")
df = df.iloc[:,1:]
df.head()
```

```python
df.info()
```

```python
# Standardize all continuous variables
ndf = pd.DataFrame(StandardScaler().fit_transform(df.iloc[:, :29]))
ndf.rename(columns = {28:"Amount"}, inplace = True)
ndf.head()
```

```python
# Column binding the continuous variables with the categorical of 0 (non-fraud) and 1 (fraud) transaction
ndf = pd.concat([ndf, df.iloc[:,29]], axis = 1)
ndf.head(5)
```

```python
# check to see continuous variables were standardized
rp.summary_cont(ndf.iloc[:, :29])
```

```python
# let us now split our data into training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(ndf.iloc[:, :29], ndf.iloc[:, 29], random_state = 100, test_size  =0.3)
train_ndf = pd.concat([X_train, y_train], axis = 1)
test_ndf = pd.concat([X_test, y_test], axis = 1)
```

```python
# Check with Logistic Regression
from sklearn.linear_model import LogisticRegression
logreg_clf = LogisticRegression()
logreg_clf.fit(train_ndf.iloc[:, :29], train_ndf.iloc[:, 29])
res = logreg_clf.predict(X_test)
```

```python
# Let us check the accuracy
# First create a simple function for this
def checkAcc(y_true, y_pred):
    a = sum(y_true == y_pred)/y_true.shape[0]
    return a
```

print("Model Accuracy is:",np.round(checkAcc(y_test, res)*100,2), "%")

```python
# Saving the true labels and predictions
a = pd.DataFrame(res)
b = pd.DataFrame(y_test)
a.reset_index(drop=True, inplace=True)
b.reset_index(drop=True, inplace=True)
results = pd.concat([b,a], axis =1)
results.rename(columns = {0: "true labels", "Class":"prediction"}, inplace = True)
results.to_csv("CCfraud.csv")
```

```python
# Save the Model for Future use
import pickle
filename = "CCfraudDet.sav"
pickle.dump(logreg_clf, open(filename, "wb"))
```

```python
# Load the model
loaded_model = pickle.load(open("CCfraudDet.sav", 'rb'))
result = loaded_model.score(X_test, y_test)
print("Loaded Model Accuracy is:", np.round(result*100,2), "%")
```

#### <font color='steelblue'>RESULT:</font> We now have created an ML model that will allow us to classify whether future transactions would be flagged as a possible fraud. The said model, as shown in the tests conducted, performs with 99.91% accuracy.
